Dataset ini berisi 79 entri data tentang produk-produk toko "KARUNGJANTAN" berasal dari platform e-commerce Tokopedia, seperti yang terlihat dari nama file. Dataset ini memiliki 14 kolom, yang sebagian besar bertipe object (string), kecuali kolom Rating Produk yang bertipe float64.

Berikut adalah beberapa kolom penting yang ada dalam dataset:

ID Order: ID Order produknya.

LINK URL: Tautan ke halaman produk.

Harga: Harga produk.

Judul Produk: Nama atau judul produk.

Rating Produk: Penilaian atau rating produk.

Terjual: Jumlah produk yang terjual.

image, image2, image3: URL gambar produk.

Kota Penjual: Kota asal penjual.

Sisa Produk: Informasi mengenai sisa stok produk.

Size Produk: Ukuran-ukuran yang tersedia untuk produk.

Harga Ongkir: Biaya pengiriman.

Total Barang: Total jumlah barang.



Dapat dilihat juga bahwa beberapa kolom seperti Terjual, image3, Sisa Produk, Size Produk, Harga Ongkir, dan Total Barang memiliki nilai yang hilang (null values), yang dimana barang tidak memiliki stok/habis.
